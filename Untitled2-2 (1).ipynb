```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import folium
palette = ["#750c6e"]
```

```python
data = pd.read_csv('https://getfile.dokpub.com/yandex/get/https://disk.yandex.ru/d/ubBDVb03J8fISg')
```

```python
display(data)
```

```python
sns.relplot(
    data=data,
    x="lat", y="lng", hue="category")
plt.title('Скретч-диаграмма всех заведений по Москве')
```

filtered_data = data[data['category'] == 'кофейня']
sns.kdeplot(
    data=filtered_data,
    x="lat", y="lng", hue="category", levels=10, fill=True,
    palette=palette)
plt.title('KDE-диаграмма кофеен по Москве')

```python
filtered_data = data[data['category'] == 'кофейня']
sns.scatterplot(
    data=filtered_data,
    x="middle_avg_bill", y="district", hue = "price")
plt.title('Проверка среднего чека по округам')
```

```python
filtered_data = data[(data['category'] == 'кофейня') & (data['price'] == 'средние')]
plt.figure(figsize=(25,15))
sns.boxplot(
    data=filtered_data,
    x="middle_avg_bill", y="district", hue="price")
plt.title('Усики')
```

```python
filtered_data = data[data['district'] == 'Западный административный округ']
fig = plt.figure(figsize = (14, 8))
sns.violinplot(
    data=filtered_data,
    x="middle_avg_bill", y="district",
    palette=palette)
plt.title('Скрипичный график который показывает кофейни относительно среднего чека в ЗАО')
```

```python
filtered_data = data[(data['district'] == 'Западный административный округ') & (data['category'] == 'кофейня')]

g = sns.relplot(
    data=filtered_data,
    x="lat", y="lng", hue="middle_avg_bill")

sns.kdeplot(
    data=filtered_data,
    x="lat", y="lng", hue="category", levels=10, fill=True, alpha=0.2, ax=g.ax)
plt.title('KDE и Скретч-диаграмма кофеен в ЗАО по среднему чеку')
```

```python
import folium
from folium.plugins import HeatMap

df = data

df["middle_avg_bill"] = pd.to_numeric(df["middle_avg_bill"])

map_data = df[["lat", "lng", "middle_avg_bill"]]

map_data.fillna(map_data.median(), inplace=True)

map = folium.Map(location=[55.7558, 37.6173], zoom_start=12)

HeatMap(map_data, radius=15).add_to(map)

map
```

```python
import plotly.express as px

url = "https://getfile.dokpub.com/yandex/get/https://disk.yandex.ru/d/ubBDVb03J8fISg"
df = pd.read_csv(url)

df["middle_avg_bill"] = pd.to_numeric(df["middle_avg_bill"])
df["seats"] = pd.to_numeric(df["seats"])

df["monthly_customers"] = df["seats"] * 1 * 30  # Предполагаем, что клиенты проводят в кофейне в среднем 1 час

# Расчет месячной прибыли
df["monthly_profit"] = df["middle_avg_bill"] * df["monthly_customers"]

profit_by_district = df.groupby("district")["monthly_profit"].sum()

fig = px.pie(
    profit_by_district,
    values="monthly_profit",
    names=profit_by_district.index,
    title="Месячная прибыль кофеен по округам",
)

fig.show()
```
